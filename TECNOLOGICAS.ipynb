```python
import requests
from bs4 import BeautifulSoup

# URL DEL SITIO WEB
url = 'https://quienesquien.laverdad.es/sector/innovacion-tecnologia-informacion.html'

try:
    # SOLICITUD WEB
    respuesta = requests.get(url, verify=False, timeout=10)
    respuesta.raise_for_status()  # Levanta un error si la solicitud devuelve un código de estado HTTP 4xx/5xx
    soup = BeautifulSoup(respuesta.content, 'html.parser')
    
   
except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error occurred: {http_err}")  # Captura errores HTTP

except requests.exceptions.RequestException as err:
    print(f"Error occurred: {err}")  # Captura cualquier otro error

except Exception as e:
    print(f"Ocurrió un error: {e}")
```

```python
contenedor = soup.find('ol', id="myList")
```
